```python
import glob
import os
```

```python
imagefiles = []
for sub_dir_path in glob.glob("/content/drive/MyDrive/out"+"*"):
    #if os.path.isdir(sub_path_dir):
    try:
        for filename in os.listdir(sub_dir_path):
          imagefiles.append(filename)
          imagefiles.sort()
    except Exception as e:
        print(e)
```

```python
imagefiles
```

```python
len(imgfiles)
```

```python
import re
```

```python
label=[]
for namefile in imagefiles:
  newname = re.sub(r'_\d+.jpg', '', namefile)
  label.append(newname)
```

```python
label
```

```python
import csv
```

```python
import csv

with open('newdata.csv', 'w', newline='') as csvfile:
    fieldnames = ['filename', 'label']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for i, j in zip(imagefiles, label):
        writer.writerow({'filename': i, 'label': j})
```
